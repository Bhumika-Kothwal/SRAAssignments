```
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def rot90(i,j):
  return j,w-i-1
def rot180(i,j):
  return h-i-1,w-j-1


image = plt.imread('rotate.png')
print(image.shape)
choice = int(input('Rotate by 1)90 Degrees\n2)180 Degrees'))
if choice == 1 :
  h = image.shape[1]
  w = image.shape[0]
elif choice == 2 :
  h = image.shape[0]
  w = image.shape[1]
   
np1 = np.zeros((h,w,image.shape[2]),dtype = image.dtype)

for s in range(4):
  for i in range(image.shape[0]):
    for j in range(image.shape[1]):
      if choice == 1 :
        x,y = rot90(i,j)
      elif choice == 2 :
        x,y = rot180(i,j)
      np1[x,y,s] = image[i,j,s]
plt.imshow(np1)

```
